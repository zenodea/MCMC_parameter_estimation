### Steps

- Italy's official government data is divided in regions. As we are looking at Italy as a whole, the data needs to be merged.
- Swab tests and Antigenic Tests are separated, thus we need to merge them as one.
- The data does not have a daily new cases of COVID-19, thus we need to get the difference per row for the cumulative positive tests.

```python
import pandas as pd
```

```python
df = pd.read_csv('../raw_data/Covid_Data/italy_covid_data.txt', sep=",")
```

```python
df = df[['Date','Dead', 'Swab Test Positive','Antigenic Test Positive']]
df['cumCases'] = df['Swab Test Positive']+df["Antigenic Test Positive"]
df = df.drop('Swab Test Positive', axis=1)
df = df.drop('Antigenic Test Positive', axis=1)
df = df.rename(columns={"Date": "date", "Dead": "cumDeath"})
df = df.groupby(['date']).sum()
df.fillna(0)
df
```

```python
df.to_csv('../processed_data/cumDeathItaly.csv')
```
