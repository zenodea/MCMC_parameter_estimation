```python
import pandas as pd
```

```python
df = pd.read_csv('../raw_data/Covid_Data/france_covid_data.csv', sep=",")
df = df[['date','dc_tot']]
df = df.fillna(0)
df['dc_tot'] = df['dc_tot'].astype(int)
df = df.rename(columns={"date": "date", "dc_tot": "cumDeath"})
df
```

```python
df.to_csv('../processed_data/cumDeathFrance.csv')
```

```python
df = pd.read_csv('../raw_data/Covid_Data/owid-covid-data.csv', sep=",")
df = df.loc[df['location'] == "France"]
df = df[['date','total_deaths']]
df = df.fillna(0)
df = df.rename(columns={"date": "date", "total_deaths": "cumDeath"})
df['cumDeath'] = df['cumDeath'].astype(int)
df
```

```python
df.to_csv('../processed_data/cumDeathFrance.csv')
```
