```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
```

```python
country_names = ['France','Germany','Italy','Spain','UK','Switzerland','Netherlands','Sweden']
nameOfFiles = "../processed_data/cumDeath"

for x in country_names:
    t = np.linspace(0, 600, 600)
    currName = nameOfFiles + x +'.csv'
    df = pd.read_csv(currName)
    deaths = df["cumDeath"].tolist()
    fig = plt.figure(facecolor='w')
    ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True, label="Deaths")
    ax.plot(df['date'], deaths)
    plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=130))
    plt.gcf().autofmt_xdate()
    plt.title(x + " Cumulative Covid-19 Deaths")
    plt.xlabel("Time / Days")
    plt.ylabel("Cumulative Deaths Due to COVID-19")
    plt.xlim(0,900)
    ax.legend(['Real World Data (Cumulative COVID-19 Deaths)','First Lockdown','Second Lockdown'])
    ax.grid(which='major', c='w', lw=2, ls='-')
```
